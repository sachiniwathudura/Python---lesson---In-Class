```python
import requests

# Base URL of the Open-Meteo API
base_url = "https://api.open-meteo.com/v1/forecast"

# Function to fetch weather data
def fetch_weather_data(latitude, longitude):
    params = {
        'latitude': latitude,
        'longitude': longitude,
        'current_weather': 'true',  # Request current weather data
    }
    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        if 'application/json' in response.headers['Content-Type']:
            return response.json()
        else:
            print("Invalid response format.")
            return None
    else:
        print("Something went wrong:", response.status_code)
        return None

# Function to parse and display weather data
def parse_weather_data(data):
    if data is None or 'current_weather' not in data:
        print("No data available.")
    else:
        current_weather = data['current_weather']
        temperature = current_weather.get('temperature', 'N/A')
        wind_speed = current_weather.get('windspeed', 'N/A')
        print("Current Temperature:", temperature, "¬∞C")
        print("Current Wind Speed:", wind_speed, "m/s")

# Fetch and parse weather data for London (latitude: 51.5074, longitude: -0.1278)
data = fetch_weather_data(51.5074, -0.1278)
parse_weather_data(data)

```
